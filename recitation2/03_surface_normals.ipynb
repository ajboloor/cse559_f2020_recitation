## Import Libraries

```python
import numpy as np
from scipy.ndimage import gaussian_filter
from scipy.signal import convolve2d as conv2
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

## Load Image

```python
img_normal = np.array(imread('true_normals.png')).astype(np.float64) / 255.
print("img shape:", img_normal.shape)
plt.figure(figsize=(12, 6))
plt.imshow(img_normal)
plt.show()
```

## Surface normal in x direction (red channel)

```python
plt.figure(figsize=(12, 6))
plt.imshow(img_normal[:, :, 0], cmap='inferno') # red channel
plt.xticks([])
plt.yticks([])
plt.title('Surface Normal in X direction')
plt.show()
```

## Surface normal in y direction (green channel)

```python
plt.figure(figsize=(12, 6))
plt.imshow(img_normal[:, :, 1], cmap='inferno') # green channel
plt.xticks([])
plt.yticks([])
plt.title('Surface Normal in Y direction')
plt.show()
```

## Surface normal in z direction (blue channel)

```python
plt.figure(figsize=(12, 6))
plt.imshow(img_normal[:, :, 2], cmap='inferno') # blue channel
plt.xticks([])
plt.yticks([])
plt.title('Surface Normal in Z direction')
plt.show()
```
